# Linguistic annotations

spaCy provides a variety of linguistic annotations to give you insights into a text’s grammatical structure.

This includes the word types, like the parts of speech, and how the words are related to each other. For example, if you’re analyzing text, it makes a huge difference whether a noun is the subject of a sentence, or the object – or whether “google” is used as a verb, or refers to the website or company in a specific context.

````python
import spacy

nlp = spacy.load("en_core_web_sm") # returns language object
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")
for token in doc:
    print(token.text, token.pos_, token.dep_)
````

Even though a Doc is processed – e.g. split into individual words and annotated – it still holds all information of the original text, like whitespace characters. 

This way, you’ll never lose any information when processing text with spaCy.

### Token Matcher

It is a token based matching given as `Matcher`, operating over tokens. It makes use of the word level features of spaCy such as `LOWER, LENGTH, LEMMA, SHAPE` and flags such as `IS_PUNCT, IS_DIGIT, LIKE_URL`, etc.

````python
#  "POS": "NOUN"
pattern_verb = [{"LOWER": 'alice', "POS":"PROPN"}, {'POS':'AUX'}]
pattern = [{"LOWER": 'alice', "POS":"PROPN"}, {'POS': {"NOT_IN": ['AUX']}}]
````

#### More Examples on Token Matching

````python
matcher = Matcher(nlp.vocab)
````

````python
pattern = [{"POS":"ADJ"}, {"POS":"NOUN"}]
````

````python
matcher.add('adj_noun', [pattern])
````

````python
matches = matcher(doc)
````

````python
for match_id, start, end in matches:
  string_id = nlp.vocab.strings[match_id]
  span = doc[start:end]
  print(match_id, string_id, start, end, span.text)

  break
````

````python
matcher = Matcher(nlp.vocab)

pattern = [{"LEMMA": "begin"}, {"POS":"ADP"}]

matcher.add('lemma_adp', [pattern])
matches = matcher(doc)
````

````python
index = 0
for match_id, start, end in (matches):
  index = index+1
  string_id = nlp.vocab.strings[match_id]
  span = doc[start:end]
  print(match_id, string_id, start, end, span.text)

  if index>10:
    break
````

````python
matcher = Matcher(nlp.vocab)

pattern = [{"TEXT": "Alice"}, {"IS_PUNCT":True, "OP":"*"}]

matcher.add('lemma_adp', [pattern])
matches = matcher(doc)
````

````python
index = 0
for match_id, start, end in (matches):
  index = index+1
  string_id = nlp.vocab.strings[match_id]
  span = doc[start:end]
  print(match_id, string_id, start, end, span.text)

  if index>10:
    break
````

### Phrase Matcher

If you need to match large terminology lists, you can also use the PhraseMatcher and create Doc objects instead of token patterns, which is much more efficient overall. The Doc patterns can contain single or multiple tokens.

````python
from spacy.matcher import PhraseMatcher
````

````python
nlp = spacy.load('en_core_web_sm')
````

````python
matcher = PhraseMatcher(nlp.vocab, attr='LOWER')
````

````python
phrase_list = ['she had', "she found"]
````

````python
pattern = [nlp.make_doc(text) for text in phrase_list]
````

````python
matcher.add("phrase_match", pattern)
````

````python
doc = nlp(text)
````

````python
matches = matcher(doc)
````

````python
for match_id, start, end in matches:
  span = doc[start:end]
  print(start, end, span.text)
````

# Entity Matcher

Here we are matching entities other than tokens or phrases. It is a matcher based on dictionary patterns and can be combined with the spaCy’s named entity recognition to make the accuracy of entity recognition much better

````python
import spacy
from spacy.pipeline import EntityRuler
from spacy.lang.en import English
````

````python
text = """The First Estate included the clergy (church leaders), the Second Estate included the nobles, and the Third Estate included the commoners.
        The Third Estate paid most of the taxes, while the nobility lived lives of luxury and got all the high-ranking jobs."""
````

````python
nlp = English()
ruler = nlp.add_pipe('entity_ruler')
````

````python
patterns = [{"label":"NOUN", "pattern":"church"}, {"label":"ORG", "pattern":[{"lower":"the"}, {"lower":{"IN": ["first", 'second', 'third']}}, {"ORTH":"Estate"}]}]
````

````python
ruler.add_patterns(patterns)
````

````python
doc = nlp(text)
````

````python
for ent in doc.ents:
  print(ent.text, ent.label_, ent.ent_id_)
````

# Word to Vector (word2vec) and Sentence Similarity

image.png

Word2vec is a technique for natural language processing published in 2013. The word2vec algorithm uses a neural network model to learn word associations from a large corpus of text. 

These models are shallow, two-layer neural networks that are trained to reconstruct linguistic contexts of words

image.png

**Detailed Description:**
https://jalammar.github.io/illustrated-word2vec/

**Tesnorflow word2vec:** https://www.tensorflow.org/tutorials/text/word2vec
